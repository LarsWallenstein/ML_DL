![alt text](https://wikimedia.org/api/rest_v1/media/math/render/svg/a242253f2365176400f50d43c9a66566132d3d92)

```
#Импортирование необходимых библиотек
import matplotlib.pyplot as plt
from scipy.stats import chi2,norm
import numpy as np


fm = 40#количество степеней свобод
vib = chi2.rvs(fm, size=1000)#генерация выборки
plt.hist(vib, 20, normed=True, histtype='stepfilled', alpha=0.5)#создание гистограммы
x = np.arange(10.,80.,0.5)
plt.plot(x, chi2.pdf(x, fm),'b', alpha=0.9, label='chi2 probability density function')#график функции плотности распределения хи-квадрат
plt.legend(loc='best', frameon=False)
plt.xlabel('x')
plt.ylabel('Probability')
plt.show()
```

```
num=[5,10,30]#Генерация выборочных средних для n=5,10,30 соответственно
tmp=[]
cells = np.arange(30.,50.,0.5)
for i in num:
  for j in np.arange(1000):
    mas=chi2.rvs(fm,size=i)
    tmp.append(np.mean(mas))
```

```
#Построение гистограммы и функции плотности нормального распределения при n=5
plt.hist(tmp[:1000], 20, normed=True, histtype='stepfilled', alpha=0.5)
plt.plot(cells,norm.pdf(cells,fm, np.sqrt(2*fm/5.)),'b', alpha=0.9, label='norm probability density function')
plt.xlabel('Sample means 5')
plt.ylabel('Probability')
plt.legend(loc='best', frameon=False)
plt.show()
```

```
#Построение гистограммы и функции плотности нормального распределения при n=10
plt.hist(tmp[1000:2000], 20, normed=True, histtype='stepfilled', alpha=0.5)
plt.plot(cells,norm.pdf(cells,fm, np.sqrt(2*fm/10.)),'b', alpha=0.9, label='norm probability density function')
plt.xlabel('Sample means 10')
plt.ylabel('Probability')
plt.legend(loc='best', frameon=False)
plt.show()
```

```
#Построение гистограммы и функции плотности нормального распределения при n=30
plt.hist(tmp[2000:3000], 20, normed=True, histtype='stepfilled', alpha=0.5)
plt.plot(cells,norm.pdf(cells,fm, np.sqrt(2*fm/30.)),'b', alpha=0.9, label='norm probability density function')
plt.xlabel('Sample means 30')
plt.ylabel('Probability')
plt.legend(loc='best', frameon=False)
plt.show()
```

Была на практике проверена ЦПТ, а именно то, что гистограмму распределения можно описать нормальым распределением с параметрами $N(\mu,\sigma^2/n)$. Также наглядно показано, что с ростом числа  n растёт и точность описания выборки с помощью нормального распределения.
