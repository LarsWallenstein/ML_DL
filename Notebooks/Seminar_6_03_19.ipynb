```
x1=[1,7,1,9]
x2=[5,9,3,7]
```

```
import matplotlib.pyplot as plt
```

```
plt.scatter(x1,x2)
```

```
x=[]
for i in range(len(x1)):
  x.append([x1[i],x2[i]])
```

```
dist =[]
for i in range(len(x)):
  for j in range(i,len(x)):
    dist.append((0.1*(x[i][0]-x[j][0])**2+0.9*((x[i][1]-x[j][1])**2))**0.5)
```

```
print dist
```

```
mi = 1000
for i in dist:
  if i<mi and i!=0 and i!=1:
    mi=i
```
